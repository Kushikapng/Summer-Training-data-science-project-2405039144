````python
#Ans_4
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.scatterplot(data=stars_df, x='Temperature (K)', y='Luminosity(L/Lo)',
                color='hotpink',
                s=150)

plt.title('Relationship between Temperature and Luminosity')
plt.xlabel('Temperature')
plt.ylabel('Luminosity');
````

The scatter plot reveals a non-linear relationship between stellar temperature and luminosity, mapping out distinct structural groupings. A dense concentration of stars forms a baseline at the lowest luminosity levels across all temperatures, representing stable main sequence and dwarf stars. In contrast, the top-left region highlights an exclusive cluster of highly luminous yet relatively cool stars, which indicates the presence of massive Red Giants.

````python
#Ans_5
sorted_stars = stars_df.sort_values(by='Radius(R/Ro)')

plt.figure(figsize=(12, 6))
plt.plot('Star type', 'Star color', data=sorted_stars, marker='*', ms= 10, c='maroon',mfc='hotpink')
plt.title('Relation of Star type with star color')
plt.xlabel('Star type')
plt.ylabel('Star color');
````

Based on the graph in image_ae9ba9.png, we can infer that certain star types have very strict color constraints, while others vary wildly. For instance, Type 0 and Type 1 stars are exclusively Red. In contrast, types like 2, 3, and 5 span a massive variety of colors, ranging from shades of Blue and White to Yellow and Orange. Additionally, the chart highlights a data cleanup issue: several identical colors are split into separate categories due to inconsistent text formatting (such as "Blue-White", "Blue white", and "Blue White"), which would need to be standardized before training a model.

````python
#Ans_6
star_counts = stars_df['Star color'].value_counts()
plt.figure(figsize=(10, 5))
plt.xticks(rotation=75)
sns.barplot(x=star_counts.index, y=star_counts, color='hotpink')
plt.title('Distribution of Star Colors')
plt.xlabel('Star Color')
plt.ylabel('Count');
````

Based on this graph, we can infer that Red stars are by far the most common in this dataset, outnumbering every other color by a wide margin. The chart also clearly highlights a major data cleanup problem: the exact same colors are being split into separate bars just because of minor differences in spelling, capitalization, or spaces (for example, "Blue-white", "Blue White", and "Blue white" are counted as completely different categories). Clean text formatting would group these together accurately.

````python
#Ans_7
class_counts = stars_df['Spectral Class'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(
    class_counts,
    labels=class_counts.index,
    autopct='%1.1f%%',
    startangle=180,
    counterclock=False,
    colors=['pink','lightcoral','crimson','mediumvioletred','palevioletred','hotpink','maroon'])


plt.title('Spectral Class Distribution');
````

Based on this pie chart, we can infer that Class M stars make up nearly half of the entire dataset at 46.2%, making them the most dominant group by far. When combined with Class B (19.2%) and Class O (16.7%), these three categories account for over 80% of all the stars listed. In stark contrast, Class G stars are extremely rare here, representing just a tiny sliver at 0.4%.

````python
#Ans_8
pastel_pinks = ['#80003a','#ff3399', '#80003a', '#b30052', '#ff3399', '#ff80bf']
sns.countplot(
    data=stars_df,
    x='Spectral Class',
    hue='Star type',
palette=pastel_pinks
)

plt.title('Distribution of Star Types within each Spectral Class', fontsize=14)
plt.xlabel('Spectral Class')
plt.ylabel('Number of Stars')
plt.legend(title='Star Type');
````

Based on this bar chart, we can infer that certain spectral classes are heavily tied to specific star types. For example, Class M is the only category that contains Type 0 and Type 1 stars, with a maximum count of 40 stars each, while also containing several Type 5 stars. Meanwhile, Class B has the most variety, containing five out of the six star types (Types 2, 3, 4, 5, and a high count of Type 2). On the other end, Class G is incredibly rare, containing only a single Type 5 star in the entire dataset.

````python
#Ans_9
matrix_data = stars_df.pivot_table(
    index="Star type",
    columns="Spectral Class",
    values="Temperature (K)",
)

plt.figure(figsize=(9, 6))
sns.heatmap(
    matrix_data,
    cmap="RdPu",
)

plt.title(
    "Distribution of Star Types by Spectral Class", fontsize=14, pad=15
)
plt.xlabel("Spectral Class", fontsize=12);
plt.ylabel("Star Type", fontsize=12);
````

we can infer that the darkest purple cells represent the highest values in the dataset, specifically occurring for Star Type 3 in Class O, Type 4 in Class B, and Type 5 in Class O. Because these intense clusters align with Class O and B stars, this grid is clearly measuring average Temperature or Luminosity. Conversely, the large white gaps highlight missing combinations where no stars exist in nature, while the faint cells in Class M highlight that Type 0 and 1 stars have the lowest values, matching the profile of cool, dim Red Dwarfs.
