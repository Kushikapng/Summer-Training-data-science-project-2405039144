```python
import pandas as pd
df = pd.read_csv("Customer-Purchase-History.csv")
print(df.head())
```

```python
#Data Cleaning
print(df.isnull().sum())

df.drop_duplicates(inplace=True)

print("Rows after cleaning:", len(df))
```

```python
#Best Selling Products
best_selling = (
    df.groupby("Product")["Quantity"]
    .sum()
    .sort_values(ascending=False)
)

print(best_selling)
```

```python
#Calculate the Monthly Revenue
monthly_sales = (
    df.groupby("Month")["TotalPrice"]
    .sum()
)

print(monthly_sales)
```

```python
#cusomers who bought the most items
customer_quantity = (
    df.groupby("CustomerName")["Quantity"]
    .sum()
    .sort_values(ascending=False)
)

print(customer_quantity.head(10))
```

```python
#revenue analysis
print("Total Revenue:", df["TotalPrice"].sum())

print("Average Revenue:", df["TotalPrice"].mean())

print("Maximum Sale:", df["TotalPrice"].max())

print("Minimum Sale:", df["TotalPrice"].min())
```
